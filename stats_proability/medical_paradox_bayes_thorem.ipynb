*This Notebook is created to illustrate why a test with 90% accuracy can still show you 9% prediction in reality.*

**Summary**

* A medical Test with 90% accuracy should not be seens as 90% accurate. Below code proves it.

* The more rare the disease the more chances are that your belief might change like below proven via Bayes Theorem.

* To calculte the accurate result for Bayes Theorem, I have used odds instead of proability as odds provides a more precise results than proability. This is also proven below.


```python
# say you have 1000 patients with covid, and say that you test is about 90% in finding the patient with positive result.

patients = 1000

# say the total number of patient with covid are 10. means 1% of the total patients have covid.
# from which 1 of them is false negitive means he doesn't have covid, but model says he has it. because of 90% accuracy of the model.

true_positive_patients = 9
false_positive_patients = 1

patients -= true_positive_patients + false_positive_patients

# also say that the test has about 9& chance of getting false positive results, from the remaining 990 patients.
false_negitives = 0.09
false_negitive_patients = int((patients * false_negitives)) # 89 false positive patients

# calculating the proability

proability_having_covid_given_positive_test = true_positive_patients / (true_positive_patients + false_negitive_patients)
print("Proability having covid given positive test", proability_having_covid_given_positive_test.__round__(2))
```

```python
# 9% is the acutal probility of having a covid instead of 90%.
# means before taking the test the proability of having a disease was 1 in 100, the result of bayes theorem now it has been updated to 1 in 11. This is known as Bayesian update.

positive_predictive_value_of_test = proability_having_covid_given_positive_test
```

```python
# *Sensitivity* means how much sensitive the test is for true values.
# *Specificity* means to check for false positives.

sensitivity = 0.9 # as 90% was the accuracy of the test.
specificity = 0.91 # say that the 

sensitivity, specificity
```

*If you want to know how much likely the strength of updating is than:*

```python
# this baes factor is also called likelyhood.
bayes_factor = (sensitivity / false_negitives).__round__(2)
bayes_factor # the test updation strength was about 10.
```

#### Calculating Bayes theorem

```python
prioir = 0.01
posterior = prioir * bayes_factor
posterior # here this is 1/10 near to 1/11.
```
